```python
# From the code above create a function or class that performs all of the steps given a database file and a streams CSV file.
import os
import pandas as pd
import numpy as np
import sqlite3
from datetime import date

def ingest_data(db_file, streams_csv):

    # 1. connect to databse
    conn = sqlite3.connect(db_file)

    # 2. read customer data
    df_customer = pd.read_sql_query("SELECT cu.customer_id, cu.last_name, cu.first_name, cu.DOB, cu.City, cu.State, " + 
                                "co.country_name, cu.gender FROM CUSTOMER as cu JOIN COUNTRY as co ON cu.country_id = co.country_id", conn)

    # 3. read invoice / subscriber type
    df_invoice = pd.read_sql_query("""
                                SELECT 
                                i.customer_id,
                                it.invoice_item AS subscriber_type
                            FROM invoice_item it
                            JOIN invoice i
                                ON it.invoice_item_id = i.invoice_item_id
                                """, conn)

    # 4. read streams CSV
    df_streams = pd.read_csv(streams_csv)

    # 5. quality checks
    #    - drop duplicates
    df_customer = df_customer.drop_duplicates(subset="customer_id", keep="first")

    #    - drop missing stream_id
    df_streams = df_streams[df_streams["stream_id"].notna()]

    # 6. merge datasets
    df_merged = pd.merge(df_customer, df_streams, on="customer_id", how="left")

    # 7. derived columns
    #    - age
    df_merged["DOB"] = pd.to_datetime(df_merged["DOB"], errors="coerce")
    df_merged["age"] = date.today().year - df_merged["DOB"].dt.year

    #    - customer_name
    df_merged["customer_name"] = df_merged["first_name"] + " " + (df_merged["last_name"])

    #    - is_subscriber
    df_merged["is_subscriber"] = df_merged["subscription_stopped"].map({
        0: False,
        1: True
    }).astype('boolean')

    # 8. count streams per customer
    streams_count = (
        df_streams.groupby("customer_id").size().reset_index(name="num_streams")
    )

    # 9. final grouping
    df_final = pd.merge(
        df_merged,
        df_invoice,
        on="customer_id",
        how="left"
    )

    df_final = pd.merge(
        df_final,
        streams_count,
        on="customer_id",
        how="left"
    )

    df_final_grouped = df_final.groupby(
        ["customer_id", "country_name", "age", "customer_name", "is_subscriber", "subscriber_type"],
        as_index=False
    ).agg(num_streams=("num_streams", "max"))

    return df_final_grouped
```

```python
data_dir = os.path.join("..", "data")
df_full = ingest_data(
    os.path.join(data_dir, "aavail-customers.db"),
    os.path.join(data_dir, "aavail-streams.csv")
)
df_full
```

```python
# Create customer ID batches

import numpy as np

batch_size = 500

customer_ids = df_full["customer_id"].unique()

batchs = np.array_split(customer_ids, len(customer_ids) // batch_size + 1)

batchs
```

```python
# Run ingestion per batch

df_batches = []

for batch_ids in batchs:
    df_batch = ingest_data(
        os.path.join(data_dir, "aavail-customers.db"),
        os.path.join(data_dir, "aavail-streams.csv")
    )

    df_batch = df_batch[df_batch["customer_id"].isin(batch_ids)]
    df_batches.append(df_batch)
```

```python
df_batched = pd.concat(df_batches, ignore_index=True)
df_batched
```

```python
sort_cols = [
    "customer_id",
    "country_name",
    "age",
    "customer_name",
    "is_subscriber",
    "subscriber_type",
    "num_streams"
]

df_full_sorted = df_full.sort_values(sort_cols).reset_index(drop=True)
df_batched_sorted = df_batched.sort_values(sort_cols).reset_index(drop=True)
df_full_sorted
```
